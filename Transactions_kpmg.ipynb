```python
df=df.dropna(axis=0, how='any',subset=None, inplace=False)
print(df.isnull().sum())
```

```python
df['profit']=df['list_price']-df['standard_cost']
df.head()
```

```python
df.drop(df[df['order_status']=='Cancelled'].index,inplace=True)
df.head()
```

```python
df.shape
```

```python
df['list_price'] = ["${:,.2f}".format(x) for x in df['list_price']]
df['standard_cost'] = ["${:,.2f}".format(x) for x in df['standard_cost']]
df['profit'] = ["${:,.2f}".format(x) for x in df['profit']]
```
